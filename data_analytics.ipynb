```python
# Analytics

if dfa.shape[0] == 0:
    print("No analytics data available yet. Showing placeholder message.")

else:
    dfa['Views'] = pd.to_numeric(dfa['Views'], errors='coerce')

    plt.figure(figsize=(10,5))

    dfa.sort_values('Views', ascending=False).head(10).plot(
        x='Page path and screen class',
        y='Views',
        kind='bar',
        legend=False
    )

    plt.title('Figure 1: Top Pages by Views')
    plt.ylabel('Views')
    plt.xlabel('Page Path and Screen Class')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
# Survey

plt.figure(figsize=(8,5))

dfs['What is your class year?'].value_counts().plot(kind='bar')

plt.title('Figure 2: Distribution of Class Years in the Survey')
plt.ylabel('Count')
plt.xlabel('Class Year')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
